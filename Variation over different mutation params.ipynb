## Analysis & Ploting 

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Data collected manually from different csv files in folder 'output\1702332572-2566428'

mutation_rates = [10, 15, 25]
max_generations = [1500, 1500, 1500]
RWS = [6601, 7473, 7265]
PS = [6927, 6572, 6371]
TS = [5694, 6251, 6285]

# Create a DataFrame
import pandas as pd
df = pd.DataFrame({'Mutation': mutation_rates, 'Max_gen': max_generations, 'Roulette': RWS, 'Pooling': PS, 'Tournament': TS})

# Melt the DataFrame to make it suitable for Seaborn
df_melted = pd.melt(df, id_vars=['Mutation', 'Max_gen'], var_name='Selection', value_name='Values')

# Plot using Seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(12, 8))
plot = sns.lineplot(x='Mutation', y='Values', hue='Selection', data=df_melted)
plt.title('Selection performance Comparison basis Mutation Rates')
plt.xlabel('Mutation Rate (%)')
plt.ylabel('Fitness of Fittest Individuals')
plt.show()
```

Pooling performs better then tournament throughout, but only at 10% mutation it performs better than roulette wheel selction
