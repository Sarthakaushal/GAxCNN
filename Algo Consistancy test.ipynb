## Analysis & Ploting 

```python
import os
import pandas as pd

data = {'Roulette Wheel':[], 'Tournament':[], 'Pooling':[]}
avg = {'Roulette Wheel':[], 'Tournament':[], 'Pooling':[]}

for fp in os.listdir("output/const-1702335127-4841661"):
    if ".json" in fp:
        continue
    df = pd.read_csv("output/const-1702335127-4841661/"+fp)
    print(fp)
    if 'roulette' in fp.lower():
        sel = 'Roulette Wheel'
    elif 'tour' in fp.lower():
        sel = 'Tournament'
    else:
        sel = 'Pooling'
    fi = df['fitness_of_FI'].loc[df['fitness_of_FI'].idxmax()]
    num_genes = df['fittest_individual_num_genes'].loc[df[
        'fittest_individual_num_genes'].idxmax()]
    print(sel ,fi, num_genes)
    data[sel].append(fi)
    avg[sel].append(num_genes)
```

```python
df_fi = pd.DataFrame(data)
df_num_genes = pd.DataFrame(avg)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import json
import pandas as pd
```

```python
with open("output/const-1702335127-4841661/time_data_Consistancy_comp_1702349560.5447245.json", 'rb') as fp:
    time_df = json.load(fp)
```

```python
time_df.keys()
```

```python
df_tim = {
    'Pooling': time_df['pool_selection'],
    'Roulette Wheel': time_df['roulette_selection'],
    'Tournament': time_df['tournament_selection']
}
df_tim = pd.DataFrame(df_tim)
df_tim = df_tim[['Roulette Wheel', 'Tournament', 'Pooling']]
```

```python
# Plot using Seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(12, 8))
plot = sns.boxplot(data=df_fi)
plt.title('Variation of Fitness of Fittest Individual over 20 runs on same population')
plt.xlabel('Selection Method')
plt.ylabel('Fitness of Fittest Individuals')
plt.show()
```

```python
# import seaborn as sns
# import matplotlib.pyplot as plt
sns.set(style="white")
# Sample data
x_values = [1, 2, 3, 4, 5]
y1_values = [10, 15, 7, 12, 8]
y2_values = [200, 150, 170, 160, 180]

# Create the first plot
sns.set(style="whitegrid")
fig, ax1 = plt.subplots(figsize=(10, 6))
fontsize = 24
color = 'tab:blue'
c = 'blue'
ax1.set_xlabel('Pooling Mechanisms', fontsize=fontsize)
ax1.set_ylabel('Fitness of Fittest Individual', color=color, fontsize=fontsize)
sns.boxplot(data=df_fi, color=color, medianprops={'color':'blue'},capprops=dict(color=c),
            whiskerprops=dict(color=c),boxprops=dict(facecolor="None"),
            flierprops=dict(color=c, markeredgecolor=c))
ax1.tick_params(axis='y', labelcolor=color, fontsize=24)
ax1.tick_params(axis='x', fontsize=24)
plt.grid(False)
# Create the second plot sharing the x-axis
ax2 = ax1.twinx()
color = 'tab:red'
c= 'red'
ax2.set_ylabel('Execution Time', color=color, fontsize=fontsize)
sns.boxplot(data=df_tim, color=color, medianprops={'color':'red'},capprops=dict(color=c),
            whiskerprops=dict(color=c),boxprops=dict(facecolor="None"),
            flierprops=dict(color=c, markeredgecolor=c))
# plt.xticks()
ax2.tick_params(axis='y', labelcolor=color, fontsize=24)
ax2.tick_params(axis='x', fontsize=24)

# Show the plot
plt.title('Fitness of Fittest individual and Variation in execution time',fontsize=24)
plt.grid(False)
plt.show()
```
